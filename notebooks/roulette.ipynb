```python
import math
from random import randrange

from jupyturtle import *
```

```python
size = 150

def triangle(radius,  angle):
    y = radius * math.sin(angle * math.pi / 180)

    right(angle)
    forward(radius)
    left(90+angle)
    forward(2*y)
    left(90+angle)
    forward(radius)
    left(180-angle)
    
def table(size):
    for i in range(8):
        triangle(size*1.2, 10)
        left(45)


def circular_spiral(radius, angle):
    turns = randrange(16, 49) / 16
    distance = 2 * math.pi * radius / 360
    for _ in range(int(turns * 360)):
        forward(distance)
        left(angle)
        radius += 0.015  # Increase the radius for each segment
        distance = 8 * math.pi * radius / 360
```

```python
make_turtle(delay=0, width=size*3, height=size*3)

lt(45/2)
table(size)

with no_pen():
    fd(size*.66)
    lt(90)
    
set_color('red')
circular_spiral(50, 1)
```
